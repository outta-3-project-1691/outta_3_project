```python
# 테스트 함수
def test_model():
    # 입출력 예시
    test_cases = [
        {
            "input": "I want to change the top of the woman to a blue sweatshirt and the bottom to a black skirt.",
            "expected_output": '{"top":["blue","sweatshirt"],"bottom":["black","skirt"]}'
        },
        {
            "input": "Change the top of the man to a red t-shirt and the bottom to blue jeans.",
            "expected_output": '{"top":["red","t-shirt"],"bottom":["blue","jeans"]}'
        },
        {
            "input": "Please switch the jacket to a green hoodie and the pants to gray shorts.",
            "expected_output": '{"top":["green","hoodie"],"bottom":["gray","shorts"]}'
        },
        {
            "input": "I want the woman to wear a yellow blouse and white trousers.",
            "expected_output": '{"top":["yellow","blouse"],"bottom":["white","trousers"]}'
        }
    ]

    # 각 테스트 케이스에 대해 모델 실행 및 결과 비교
    for i, case in enumerate(test_cases):
        user_input = case["input"]
        expected_output = case["expected_output"]

        # 모델 결과 생성
        generated_output = LLM(user_input)

        # 결과 비교
        if generated_output.replace('\n', '') == expected_output:
            print(f"Test Case {i + 1}: Passed")
        else:
            print(f"Test Case {i + 1}: Failed")
            print(f"  Expected: {expected_output}")
            print(f"  Generated: {generated_output}")
```

```python
# 테스트 실행
test_model()
```
